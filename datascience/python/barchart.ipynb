```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
prodi = ('Teknik Informatika (S1)','Rekayasa Perangkat Lunak (S1)','Sistem Informasi (S1)',
         'Manajemen Informatika (D3)','Komputerisasi Akuntansi (D3)')
studentbody=(317,36,55,218,166)
```

```python
x_coords=np.arange(len(prodi))
plt.bar(x_coords,studentbody)
plt.show()
```

```python
x_coords=np.arange(len(prodi))
plt.bar(x_coords, studentbody,tick_label=prodi)
plt.xticks(rotation=90) #rotates text for x-axis label
plt.show()
```

```python
x_coords=np.arange(len(prodi))
plt.bar(x_coords, studentbody,tick_label=prodi)
plt.xticks(rotation=90) #rotates text for x-axis label
plt.ylabel('Student Body (Ratusan)')
plt.title('Data Statistik Mahasiswa STMIK IKMI CIREBON')
plt.show()
```
