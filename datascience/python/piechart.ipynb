```python
import matplotlib.pyplot as plt
```

```python
flavors=('Chocolate','Vanilla','Pistachio','Mango','Strawberry')
votes=(12,11,4,8,7)
plt.title('Favorite Ice Cream Flavors')
plt.pie(votes,
       labels=flavors,
       explode=(0, 0.2,0,0,0.1),
       radius=1,#Radius Lingkaran,
        autopct='%.1f%%',
        shadow=True,
        startangle=180,#0:Horizontal kanan,
        colors=('#8B4513','#FFF8DC','#93C572','#E67F0D','#D53032'))
plt.show()
```
